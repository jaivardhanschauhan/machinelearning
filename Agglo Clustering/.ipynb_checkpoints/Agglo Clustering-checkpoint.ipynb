```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
import scipy.cluster.hierarchy as sch
```

The data set concerns the earliest history of mankind. Prehistoric men created the desired shape of a stone tool
by striking on a raw stone, thus splitting off flakes, the waste products of the crafting process. Archaelogists do not find many tools, but they do find flakes. The data set is about these flakes. Each row stands for whole inventories of these flakes. 

First Column refers to Identifier/ID of the Inventory. These identifiers are used to uniquely identify each inventory. These IDs could be further grouped based on criteria defined by archaeologists. These criteria may include factors such as age and hominid type. 

Numeric features:

LBI: Length-breadth index of the striking platform

RTI: Relative-thickness index of the striking platform

WDI: Width-depth index of the striking platform

FLA: Flaking angle (the angle between the striking platform and the splitting surface)

PSF: platform primery (yes/no, relative frequency)

FSF: Platform facetted (yes/no, relative frequency)

ZDF1: Dorsal surface totally worked (yes/no, relative frequency)

PROZD: Proportion of worked dorsal surface (continuous)


LBI, RTI, WDI, FLA, and PROZD are averages, PSF, FSF, and ZDF1 are relative frequencies.

```python
df = pd.read_csv('StoneFlakes.csv')
print(df.shape)
df.head()
```

```python
df_ac = df.copy() # Creating a new dataframe for clustering
```

```python
# Drop the non-numeric column
df_ac.drop(columns='ID',inplace=True)
```

```python
# Standardization
scaler = StandardScaler()
df_ac = scaler.fit_transform(df_ac)
```

```python
# hierarchical clustering 
linkage_matrix = sch.linkage(df_ac, method='complete')
```

```python
# plot dendogram
plt.figure(figsize=(20, 12))
dendrogram = sch.dendrogram(linkage_matrix)
plt.ylabel('height')
plt.savefig('dendogram.png', dpi=300)
```

```python
plt.figure(figsize=(20, 12))
dendrogram = sch.dendrogram(linkage_matrix)

# horizontal line at the height where there are 5 clusters
plt.axhline(y=5.75, color='red', linestyle='--', linewidth=2)  # Adjust the y-coordinate as needed

plt.ylabel('Height')
plt.title('Dendrogram with Horizontal Line at 5 Clusters')
plt.savefig('dendrogram_with_horizontal_line.png', dpi=300)
plt.show()
```

```python
cluster = AgglomerativeClustering(n_clusters=5, linkage='complete')
```

```python
cluster_preds = cluster.fit_predict(df_ac)
print(cluster_preds)
```

```python
# Attempt at visualizing the clusters in two dimensions

import numpy as np
plt.figure(figsize=(10, 6))

# Assigning colors to each cluster group
colors = ['red', 'blue', 'green', 'purple', 'orange']  

# Scatter plot for each cluster group
for cluster_id in np.unique(cluster_preds):
    plt.scatter(df_ac[cluster_preds == cluster_id, 1], df_ac[cluster_preds == cluster_id, 5], color=colors[cluster_id], label=f'Cluster {cluster_id}')

plt.xlabel('RTI (Relative Thickness Index)') # Selected 1 average measure
plt.ylabel('FSF (Flake Scar Features)') # Selected 1 relative frequency measure 
plt.title('Cluster Assignments')
plt.legend()
plt.savefig('scatter_plot_with_clusters.png', dpi=300)
plt.show()
```

We have succesfully predicted cluster label for all our data points.

```python
df['Cluster_Label'] = cluster_preds
```

```python
df.head()
```

```python
clusters = df.groupby('Cluster_Label')['ID'].apply(list)
```

```python
for cluster_label, ids in clusters.items():
    print(f"Cluster {cluster_label}: {ids}")
```
