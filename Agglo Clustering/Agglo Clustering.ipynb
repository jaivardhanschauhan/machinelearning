```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
import scipy.cluster.hierarchy as sch
```

```python
df = pd.read_csv('/content/drive/MyDrive/MSIM/datasets/StoneFlakes.csv')
df
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Drop the non-numeric column
df.drop(columns='ID',inplace=True)
```

```python
scaler = StandardScaler()
df = scaler.fit_transform(df)
```

```python
linkage_matrix = sch.linkage(df, method='complete')
```

```python
# plot dendogram
dendrogram = sch.dendrogram(linkage_matrix)
plt.ylabel('height')
plt.savefig('dendogram.png', dpi=300)
```

```python
cluster = AgglomerativeClustering(n_clusters=5, linkage='complete')
```

```python
cluster_preds = cluster.fit_predict(df)
print(cluster_preds)
```
